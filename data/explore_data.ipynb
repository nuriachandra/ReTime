```python
# import pyarrow as pa
# import pyarrow.dataset as ds

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Path to the Arrow file
arrow_file_path = "kernelsynth-data.arrow"  # Adjust if your file is elsewhere


# Visualize a few time series
def visualize_time_series(data, num_series=4):
    plt.figure(figsize=(12, 8))

    if isinstance(data, pd.DataFrame):
        # If data is a DataFrame from pandas
        for i in range(min(num_series, len(data))):
            plt.subplot(num_series, 1, i + 1)
            ts = data.iloc[i]["target"]
            plt.plot(ts)
            plt.title(f"Time Series {i + 1}")
    else:
        # If data is from GluonTS
        for i, entry in enumerate(list(data)[:num_series]):
            plt.subplot(num_series, 1, i + 1)
            plt.plot(entry["target"])
            plt.title(f"Time Series {i + 1}")

    plt.tight_layout()
    plt.show()
```
